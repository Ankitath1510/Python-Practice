```python
import pandas as pd

data ={
    'Name': ['Foo', 'Boo' ,'Hoo'],
    'Science': [95, 65, 85],
    'Mathematics': [70, 90, 80]
}

df = pd.DataFrame(data)
df
```

```python
print(data)
```

```python
df['Total']=df['Science']+df['Mathematics']
```

```python
df
```

```python
df['Average']=df['Total']/2
```

```python
df
```

```python
print(df['Average']>80)
```

```python
df[df['Average']>80]
```

```python
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Total'])
plt.title("Total marks of students")
plt.xlabel("Names")
plt.ylabel("Marks")
```
